# Summary

```python
from pandas import read_csv
analysis = read_csv("../results/results.csv",index_col=[0])
```

```python
analysis = analysis.reindex(analysis.mean(axis=1).sort_values(ascending=False).index, axis=0)
```

## Summed explained intertia

```python
analysis.T.cumsum().plot(title="Sum of explained intertia",figsize=(15,10))
pass
```

## Explained intertia for components

```python
analysis.T.plot.bar(title="Explained intertia",figsize=(15,10))
pass
```

## Fit time results

```python
fit_time = read_csv("../results/fit_time_results.csv",index_col=[0])
fit_time.T.plot(title="Fiting efficiency",figsize=(15,10))
pass
```

## Transform time results

```python
tranform_time = read_csv("../results/tranform_time_results.csv",index_col=[0])
tranform_time.T.plot(title="Transforming efficiency",figsize=(15,10))
pass
```

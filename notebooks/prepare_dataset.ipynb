```python
import pandas as pd
```

## Read dataset

```python
customer_churn = pd.read_csv('../data/Customer_churn.csv', index_col="customerID")
customer_churn.head()
```

## Discretize data for tests

```python
from jenkspy import jenks_breaks
def discretize(df,column,nb_class,label="bucket_"):
    breaks = jenks_breaks(df[column], nb_class=nb_class)
    labels = [label+str(i) for i in range(1,nb_class+1)]
    df[column]=pd.cut(df[column],
                        bins=breaks,
                        labels=labels,
                        include_lowest=True)


customer_churn['SeniorCitizen'].replace({0: 'No', 1: 'Yes'}, inplace=True)
discretize(customer_churn,'tenure',4)
discretize(customer_churn,'MonthlyCharges',4)
customer_churn['TotalCharges']= pd.to_numeric(customer_churn['TotalCharges'],errors='coerce')
customer_churn = customer_churn.dropna()
discretize(customer_churn,'TotalCharges',5)
customer_churn = customer_churn.apply(lambda x: x.astype("category"),axis=0)
customer_churn.info()
```
